```python
# Import uuid library to generate unique image names
cap = cv2.VideoCapture(0)
while cap.isOpened():
    returnvalue, frame = cap.read()
    cv2.imshow('Image', frame)
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
